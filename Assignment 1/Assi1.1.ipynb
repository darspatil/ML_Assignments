```python
# Converting Variable Data Types
for col in data.select_dtypes(include=['float64']).columns:
    data[col] = data[col].astype('float32')  # Converting all float64 columns to float32

for col in data.select_dtypes(include=['int64']).columns:
    data[col] = data[col].astype('int32')

print("Updated Data Types After Converting Float64 to Float32 and int64 to int32:")
print(data.dtypes)
```
